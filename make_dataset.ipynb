```python
import os
import pandas as pd 
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
data=[]
colors=[]
```

```python
path="C:/Users/charan/Downloads/archive/training_dataset"
  
```

```python
for dir in os.listdir(path):
    for img in os.listdir(path+'/'+dir):
        image=cv2.imread(path+'/'+dir+'/'+img)
        data.append(image[0][0])
        colors.append(dir)
        print(image[0][0],"  ",dir)
```

```python
df = pd.DataFrame(data, columns=['b', 'g', 'r'])
df["color"]=colors
df
```

```python
df.to_csv('colors.csv', index=False)
```
